```python
from lexico import *
from sintactico_ast import *
import json

# 1. Simulamos los tokens que te daría tu analizador léxico para el siguiente código C#:
# int main() {
#     int x = 15;
#     Console.WriteLine(x);
#     return 0;
# }

tokens_prueba = [
    ('KEYWORD', 'int'), ('IDENTIFIER', 'main'), ('DELIMITER', '('), ('DELIMITER', ')'), ('DELIMITER', '{'),
    ('KEYWORD', 'int'), ('IDENTIFIER', 'x'), ('OPERATOR', '='), ('NUMBER', 15), ('DELIMITER', ';'),
    ('IDENTIFIER', 'Console'), ('DELIMITER', '.'), ('IDENTIFIER', 'WriteLine'), ('DELIMITER', '('), ('IDENTIFIER', 'x'), ('DELIMITER', ')'), ('DELIMITER', ';'),
    ('KEYWORD', 'return'), ('NUMBER', 0), ('DELIMITER', ';'),
    ('DELIMITER', '}')
]

# 2. Instanciamos tu Parser con los tokens y generamos el AST
parser = Parser(tokens_prueba)
arbol = parser.parsear()

# 3. Llamamos al método que construimos para traducir a Ensamblador NASM
codigo_ensamblador = arbol.generarCodigo()

# 4. Creamos y escribimos el archivo físico en la carpeta actual
nombre_archivo = "programa.asm"
with open(nombre_archivo, "w") as archivo:
    archivo.write(codigo_ensamblador)

print(f"¡Éxito! El archivo {nombre_archivo} ha sido generado.")
```
